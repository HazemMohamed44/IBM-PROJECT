```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore", category=FutureWarning)
import matplotlib.pyplot as plt

def make_graph(stock_data, revenue_data, stock):
    stock_data_specific = stock_data[stock_data.Date <= '2021-06-14']
    revenue_data_specific = revenue_data[revenue_data.Date <= '2021-04-30']

    fig, axes = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

    # Stock price
    axes[0].plot(pd.to_datetime(stock_data_specific.Date), stock_data_specific.Close.astype("float"), label="Share Price", color="blue")
    axes[0].set_ylabel("Price ($US)")
    axes[0].set_title(f"{stock} - Historical Share Price")

    # Revenue
    axes[1].plot(pd.to_datetime(revenue_data_specific.Date), revenue_data_specific.Revenue.astype("float"), label="Revenue", color="green")
    axes[1].set_ylabel("Revenue ($US Millions)")
    axes[1].set_xlabel("Date")
    axes[1].set_title(f"{stock} - Historical Revenue")

    plt.tight_layout()
    plt.show()
    
```

```python
import requests  # مكتبة لإرسال طلبات HTTP وجلب صفحات الويب
import pandas as pd  # مكتبة لمعالجة البيانات في شكل جداول (DataFrame)
from bs4 import BeautifulSoup  # مكتبة لتحليل صفحات HTML واستخراج البيانات منها
import matplotlib.pyplot as plt  # مكتبة لرسم الجرافات
import yfinance as yf  # مكتبة لجلب بيانات الأسهم من Yahoo Finance
import warnings  # مكتبة للتحكم في التحذيرات التي تظهر في الكود

warnings.filterwarnings("ignore")  # تجاهل التحذيرات حتى لا تظهر أثناء تشغيل الكود


# -------- Function to plot graph --------
def make_graph(stock_data, revenue_data, stock):  # تعريف دالة ترسم الجراف وتستقبل بيانات السهم والإيرادات واسم الشركة
    
    stock_data_specific = stock_data[stock_data.Date <= '2021-06-14']  
    # اختيار بيانات السهم حتى تاريخ 14 يونيو 2021 فقط

    revenue_data_specific = revenue_data[revenue_data.Date <= '2021-04-30']  
    # اختيار بيانات الإيرادات حتى تاريخ 30 أبريل 2021

    fig, axes = plt.subplots(2, 1, figsize=(12,8), sharex=True)  
    # إنشاء شكل فيه جرافين فوق بعض (2 صف و1 عمود)


    # -------- رسم سعر السهم --------
    axes[0].plot(pd.to_datetime(stock_data_specific.Date),   # تحويل التاريخ إلى صيغة تاريخ صحيحة
                 stock_data_specific.Close.astype("float"),  # تحويل سعر الإغلاق إلى رقم عشري
                 label="Share Price")  # اسم الخط في الجراف

    axes[0].set_ylabel("Price ($US)")  # كتابة اسم المحور الرأسي
    axes[0].set_title(stock + " Historical Share Price")  # عنوان الجراف


    # -------- رسم الإيرادات --------
    axes[1].plot(pd.to_datetime(revenue_data_specific.Date),  # تحويل التاريخ
                 revenue_data_specific.Revenue.astype("float"),  # تحويل الإيرادات إلى أرقام
                 label="Revenue")

    axes[1].set_ylabel("Revenue ($US Millions)")  # اسم المحور الرأسي
    axes[1].set_xlabel("Date")  # اسم المحور الأفقي
    axes[1].set_title(stock + " Historical Revenue")  # عنوان الجراف

    plt.tight_layout()  # ترتيب الجرافات بحيث لا تتداخل
    plt.show()  # عرض الجراف على الشاشة



# -------- استخراج بيانات سهم Tesla --------
tesla = yf.Ticker("TSLA")  
# إنشاء كائن يمثل سهم Tesla باستخدام مكتبة yfinance

tesla_data = tesla.history(period="max")  
# جلب كل البيانات التاريخية للسهم

tesla_data.reset_index(inplace=True)  
# إعادة ترتيب الجدول بحيث يصبح التاريخ عمود عادي



# -------- Web Scraping لاستخراج الإيرادات --------
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"  
# رابط الصفحة التي تحتوي على جدول الإيرادات

html_data = requests.get(url).text  
# إرسال طلب للموقع والحصول على كود الصفحة HTML


soup = BeautifulSoup(html_data, "html.parser")  
# تحليل صفحة HTML حتى نستطيع استخراج البيانات منها


tesla_revenue = pd.DataFrame(columns=["Date","Revenue"])  
# إنشاء جدول فارغ فيه عمودين: التاريخ والإيرادات


for row in soup.find_all("tbody")[1].find_all("tr"):  
# البحث عن صفوف الجدول داخل الصفحة

    col = row.find_all("td")  
    # استخراج الأعمدة داخل كل صف

    if len(col) != 0:  
        # التأكد أن الصف يحتوي على بيانات

        date = col[0].text  
        # استخراج التاريخ من العمود الأول

        revenue = col[1].text  
        # استخراج الإيرادات من العمود الثاني

        tesla_revenue = pd.concat(
            [tesla_revenue,
             pd.DataFrame({"Date":[date],"Revenue":[revenue]})],
            ignore_index=True
        )
        # إضافة الصف الجديد إلى جدول الإيرادات



# -------- تنظيف البيانات --------
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(r",|\$", "", regex=True)  
# إزالة علامة الدولار $ والفاصلة , حتى تصبح الأرقام صالحة للحساب

tesla_revenue.dropna(inplace=True)  
# حذف الصفوف الفارغة

tesla_revenue = tesla_revenue[tesla_revenue["Revenue"] != ""]  
# حذف الصفوف التي لا تحتوي على قيمة إيرادات



tesla_revenue.tail()  
# عرض آخر 5 صفوف من جدول الإيرادات للتأكد أن البيانات صحيحة



# -------- رسم الجراف --------
make_graph(tesla_data, tesla_revenue, "Tesla")  
# استدعاء الدالة لرسم جراف سعر السهم والإيرادات لشركة Tesla
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Download the webpage
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(url).text

# Step 2: Parse the HTML data using BeautifulSoup
soup = BeautifulSoup(html_data, "html.parser")

# Step 3: Extract the Tesla Revenue table into a dataframe
# Find all tables and locate the one with "Tesla Revenue"
tables = soup.find_all("table")

tesla_revenue = pd.DataFrame(columns=["Date", "Revenue"])

for table in tables:
    if "Tesla Quarterly Revenue" in str(table):
        for row in table.find("tbody").find_all("tr"):
            cols = row.find_all("td")
            if len(cols) >= 2:
                date = cols[0].text.strip()
                revenue = cols[1].text.strip()
                tesla_revenue = pd.concat(
                    [tesla_revenue, pd.DataFrame({"Date": [date], "Revenue": [revenue]})],
                    ignore_index=True
                )
        break

# Step 4: Remove comma and dollar sign from the Revenue column
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$', "", regex=True)

# Step 5: Remove null or empty strings in the Revenue column
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]

# Step 6: Display the last 5 rows
tesla_revenue.tail()
```
